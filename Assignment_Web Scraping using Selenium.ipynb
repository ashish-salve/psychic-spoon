#### Q.2 Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name,experience_required. You have to scrape first 10 jobs data.

#### Please note that you have to scrape full job description. 

```python
# Let's first connect to the webdriver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
# opening the website
url="https://www.naukri.com/"
driver.get(url)
```

```python
# getting the job search bar web-element
job_search=driver.find_element_by_id("qsb-keyword-sugg")

# writing on job search bar
job_search.send_keys("Data Analyst")

# getting the location search bar web element
loc_search=driver.find_element_by_id("qsb-location-sugg")

# writing on location search bar
loc_search.send_keys("Bangalore")

#getting the search button web element
search_btn=driver.find_element_by_xpath("//button[@class='btn']")

#clicking the search button
search_btn.click()
```

```python
# get job title tags
title_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
#get job title text from above tags
titles=[]
titles.clear()
for i in title_tags:
    titles.append(i.text)

main_window=driver.current_window_handle
len(titles)
```

```python
companies=[]
companies.clear()

locations=[]
locations.clear()

exp=[]
exp.clear()

descriptions=[]
descriptions.clear()

jobs=pd.DataFrame({})

for i in range(0,len(titles)):
    #storing the webelement for [i]th job description
    temp_tag=title_tags[i]
    #clicking on the [i]th job description
    temp_tag.click()
    
    #storing the window handle of new tab
    child_win=driver.window_handles[i+1]
    
    # working on child window webelements
    driver.switch_to.window(child_win)
    
     # get company name tag
    comp_tag=driver.find_element_by_xpath("//*[@class='jd-header-comp-name' or @class='cpName f14']")
    #get company name text from above tags
    companies.append(comp_tag.text)
    
    #get location tags
    location_tag=driver.find_element_by_xpath("//*[@class='loc' or @class='cityLinks']")
    #get location text from above tag
    locations.append(location_tag.text)
    
    # get experience tags
    experience_tag = driver.find_element_by_xpath("//*[@class='exp' or @class='slide-meta-exp pull-left']")
    #get experience text from above
    exp.append(experience_tag.text)
    
    # get description tag
    descr_tag=driver.find_element_by_xpath("//*[@class='dang-inner-html' or @class='clearboth description']")
    #get description text from above
    descriptions.append(descr_tag.text)
    
    driver.switch_to.window(main_window)
    del temp_tag
    
#enter data into dataframe
jobs['titles']=titles
jobs['company']=companies
jobs['Location']=locations
jobs['Experience']=exp
jobs['Job Description']=descriptions
```

```python
# let's connect to thewebdriver
driver=webdriver.Chrome('chromedriver.exe')

# maximize
driver.maximize_window()

# open the glassdoor url in driver
url="https://www.glassdoor.co.in/Salaries/index.htm"
driver.get(url)

# get sign-in button web element
sign_in_btn=driver.find_element_by_xpath("//button[@class='d-none d-lg-block p-0 LockedHomeHeaderStyles__signInButton']")
# click on the sign-in button
sign_in_btn.click()


# get web-element for username bar
user_name=driver.find_element_by_id('userEmail')
# write the username
user_name.send_keys("ashish7@sharklasers.com")
# get web-element for password bar:
password=driver.find_element_by_id('userPassword')
# write the password
password.send_keys("8208507760")
# get webelement for sign-in button
sign=driver.find_element_by_xpath("//button[@class='gd-ui-button minWidthBtn css-8i7bc2']")
# click on the sign button
sign.click()

# add timer to give driver time to load the webpage
time.sleep(2)

#get the job search bar webelement
job_search=driver.find_element_by_xpath("//input[@name='sc.keyword']")
# write the text into search bar
job_search.send_keys("Data Scientist")

# get the location search bar web-element
loc_search=driver.find_element_by_id("LocationSearch")
# remove the existing text
loc_search.send_keys(Keys.CONTROL+'a',Keys.BACKSPACE)
# write the text into location search bar
loc_search.send_keys("Noida")

# get webelement for search button
search_btn=driver.find_element_by_xpath("//button[@id='HeroSearchButton']")
# click on the search button
search_btn.click()

# add timer to give driver time to load the webpage
time.sleep(2)

# get web element for job comp name
comp_wbe=driver.find_elements_by_xpath("//div[@data-test='job-info']")
# get text from above webelements
comps=[]
comps.clear()
temp_comps=[]
temp_comps.clear()
for i in comp_wbe:
    temp_comps.append(i.text)
    if len(temp_comps)==10:
        break
        
for i in temp_comps:
    comps.append(i.replace("\n",',').split(',')[1])

# get min max salary webelement
minmax_wbe=driver.find_elements_by_xpath("//div[@class='common__RangeBarStyle__rangeBar undefined undefined ']")
#get text from above webelement
sal_range=[]
sal_range.clear()
for i in minmax_wbe:
    sal_range.append(i.text)
    if len(sal_range)==10:
        break

min_sal=[]
max_sal=[]
min_sal.clear()
max_sal.clear()
for i in sal_range:
    min_sal.append(i.replace("\n"," ").split(" ")[0])
    max_sal.append(i.replace("\n"," ").split(" ")[1])

    
    
#get webelement for average salary
avg_sal_wbe=driver.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']")
# get text from above
avg_sal=[]
avg_sal.clear()
for i in avg_sal_wbe:
    avg_sal.append(i.text)
    if len(avg_sal)==10:
        break

avg_sal_1=[]
avg_sal_1.clear()
for i in avg_sal:
    avg_sal_1.append(i.replace("\n"," "))

#driver.quit()

# create dataframe
job_4=pd.DataFrame({})
job_4['Companies']=comps
job_4['Min Salary']=min_sal
job_4['Max Salary']=max_sal
job_4['Average Salary']=avg_sal_1
```

## Q.6 Scrape data of first 100 sunglasses listings on flipkart.com

```python
def fetch_data():
    # get web element for brand name
    brand_wbe=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    #get brand name text
    for i in brand_wbe:
        brands.append(i.text)
        if len(brands)==100:
            break
        
    # get web element for Price
    price_wbe=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    # get price text from above
    for i in price_wbe:
        prices.append(i.text)
        if len(prices)==100:
            break

    #get web element for discount
    disc_wbe=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
    # get text from above
    for i in disc_wbe:
        discounts.append(i.text)
        if len(discounts)==100:
            break

    #get web element for product description
    descr_wbe=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    # get text from above
    for i in descr_wbe:
        descriptions.append(i.text)
        if len(descriptions)==100:
            break
```

### Q.9 - Scraping myntra website for black shoes with price filtering

```python
def fetch_data():
    # get webelement for brand name
    brand_wbe=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h3")
    #get text
    for i in brand_wbe:
        brands.append(i.text)
        if len(brands)==100:
            break
    
        
    #get webelement for product description
    descr_wbe=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h4[@class='product-product']")
    # get text
    for i in descr_wbe:
        descriptions.append(i.text)
        if len(descriptions)==100:
            break
    
    #get webelement for price
    price_wbe=driver.find_elements_by_xpath("//div[@class='product-price']/span[1]")
    #get text
    for i in price_wbe:
        if len(prices)==100:
            break
        else:
            temp_text=i.text
            if len(temp_text)<9:
                prices.append(temp_text)
            else:
                temp_text2=temp_text.split('Rs. ')[1]
                prices.append("Rs. "+temp_text2)
```
