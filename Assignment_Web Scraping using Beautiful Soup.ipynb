### Q1.  Write a python pgm to display all the header tags from 'en.wikipedia.org/wiki/Main_Page'

```python
def wiki_headers(link):
    """function to capture header tags available from wikipedia main page"""
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # Below block will fetch the header tags
    headers=soup.find_all('h2',class_="mp-h2")
    headers_list=[]
    for i in headers:
        headers_list.append(i.text)
        
    # Below block will create dataframe
    wiki_headers=pd.DataFrame({})
    wiki_headers["Today's Headings"]=headers_list
    
    return wiki_headers
```

### Q2. Write a python program to display IMDB's Top Rated 100 movies' data(i.e Name,IMDB Rating,Year of Release) and save it in form of a csv file

```python
def imdb_100(link):
    """function gives IMDB top 100 movie name, movie rating and movie release year"""
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # below block of code is to fetch movie names
    
    m_name=soup.find_all('h3',class_='lister-item-header')
    movie_name=[]
    for i in m_name:
        movie_name.append(i.text.replace('\n',''))
    for i in range(len(movie_name)):
        movie_name[i]=movie_name[i].split('.')[1]
        
        
    # below block of code is to fetch movie ratings
    m_rate=soup.find_all('span',class_="ipl-rating-star__rating")
    temp_list=[]
    for i in m_rate:
        temp_list.append(i.text)
    movie_ratings=[]
    for i in range(0,2300):
        if i%23==0:
            movie_ratings.append(temp_list[i])
            
    # Below block of code fetches release year of movies
    m_year=soup.find_all('span',class_='lister-item-year text-muted unbold')
    movie_year=[]
    for i in m_year:
        movie_year.append(i.text)
    for i in range(len(movie_year)):
        movie_year[i]=movie_year[i].replace("(","").replace(")","")
        
    # Below block of code creates a dataframe
    top_100=pd.DataFrame({})
    top_100['Movie Name']=movie_name
    top_100['Rating']=movie_ratings
    top_100['Release Year']=movie_year
    
    #upload as csv
    top_100.to_csv("Top 100 Movies.csv")
    
    return top_100
```

```python
imdb_100('https://www.imdb.com/list/ls055592025/')
```

### Q.4 Write a python program to scrap book name,author name,genre and book review of any 5 books from "www.bookpage.com"

```python
def book_info(link):
    """function gives book name,author name,genre,book review"""
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # below block fetches book names
    temp1=soup.find_all('h4',class_='italic')
    book_names=[]
    for i in temp1:
        book_names.append(i.text.replace('\n',''))
    
    # below block fetches author names
    temp2=soup.find_all('p',class_='sans bold')
    author_names=[]
    for i in temp2:
        author_names.append(i.text.replace('\n',''))
        
    # below block fetches genre
    temp3=soup.find_all('p',class_='genre-links hidden-phone')
    genres=[]
    for i in temp3:
        genres.append(i.text.replace('\n',''))
        
    # Below block fetches reviews
    temp4=soup.find_all('p',class_='excerpt')
    reviews=[]
    for i in temp4:
        reviews.append(i.text.replace('\n',''))
        
    # below block creates dataframe
    book=pd.DataFrame({})
    book['Book']=book_names
    book['Author']=author_names
    book['Genre']=genres
    book['Reviews']=reviews
    
    return book
```

### Q.5 Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. 

### You have to scrape:

###            Q.5.1. 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def top10_teams(link):
    """Top 10 ODI Teams in men cricket"""
    # below block will get page content
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # Below block will get top 10 team names
    teams=[]
    teams.clear()
    temp_teams1=soup.find_all('span',class_='u-hide-phablet')
    for i in temp_teams1:
        teams.append(i.text)
    #teams[0:10]
    
    # Below block will fetch matches
    matches1=[]
    matches1.clear()
    temp_matches1=soup.find_all('td',class_='rankings-block__banner--matches')
    for i in temp_matches1:
        matches1.append(i.text)
    
    temp_matches2=soup.find_all('td',class_='table-body__cell u-center-text')
    temp_matches3=[]
    for i in temp_matches2:
        temp_matches3.append(i.text)
    for i in range(len(temp_matches3)):
        if i%2==0:
            matches1.append(temp_matches3[i])
    
    #Below block will fetch points
    temp_points=[]
    temp_points.clear()
    temp_points1=soup.find('td',class_='rankings-block__banner--points')
    temp_points.append(temp_points1.text)
    
    for i in range(len(temp_matches3)):
        if i%2!=0:
            temp_points.append(temp_matches3[i])
        
    # Below block will fetch ratings
    temp_ratings=[]
    temp_ratings.clear()
    temp_ratings1=soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
    for i in temp_ratings1:
        temp_ratings.append(i.text.replace('\n','').strip())
    
    temp_ratings2=soup.find_all('td',class_='table-body__cell u-text-right rating')
    for i in temp_ratings2:
        temp_ratings.append(i.text)
    
    
    # Below block creates DataFrame
    top_teams=pd.DataFrame({})
    top_teams['Team Name']=teams[0:10]
    top_teams['Matches']= matches1
    top_teams['Points']=temp_points
    top_teams['Ratings']=temp_ratings
    
    
    return top_teams
```

### Q.6 Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

### Q.6.1. Top 10 ODI teams in women’s cricket along with the records for matches, points and rating. 

```python
def top_teams_women(link):
    """This function gives the top 10 ODI teams in women cricket"""
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # Below block of code will fetch team names
    temp_teams1=soup.find_all('span',class_='u-hide-phablet')
    teams=[]
    for i in temp_teams1:
        if i.text!='': 
            teams.append(i.text)
    
    # Below block of code will fetch matches
    temp_matches1=soup.find_all('td',class_='rankings-block__banner--matches')
    matches1=[]
    for i in temp_matches1:
        matches1.append(i.text)
    
    temp_matches2=soup.find_all('td',class_='table-body__cell u-center-text')
    temp_matches3=[]
    for i in temp_matches2:
        temp_matches3.append(i.text)
    for i in range(len(temp_matches3)):
        if i%2==0:
            matches1.append(temp_matches3[i])

    # Below block of code will fetch points
    temp_points=[]
    temp_points.clear()
    temp_points1=soup.find('td',class_='rankings-block__banner--points')
    temp_points.append(temp_points1.text)
    
    for i in range(len(temp_matches3)):
        if i%2!=0:
            temp_points.append(temp_matches3[i])

    # Below block of code will fetch ratings
    temp_ratings=[]
    temp_ratings.clear()
    temp_ratings1=soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
    for i in temp_ratings1:
        temp_ratings.append(i.text.replace('\n','').strip())
    
    temp_ratings2=soup.find_all('td',class_='table-body__cell u-text-right rating')
    for i in temp_ratings2:
        temp_ratings.append(i.text)

    
    
    
    # Below block of code will create DataFrame
    top_team=pd.DataFrame({})
    top_team['Name']=teams
    top_team['Matches']=matches1
    top_team['Points']=temp_points
    top_team['Ratings']=temp_ratings
    
    
    return top_team
```

### Q.7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def amz_mobile_below_20k(link):
    """This function gives mobiles below Rs.20K from Amazon.in"""
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # Below block of code will fetch the product names
    temp_name1=soup.find_all('a',class_="a-link-normal a-text-normal")
    temp_name2=[]
    temp_name2.clear()
    for i in temp_name1:
        temp_name2.append(i.text.split("|")[0])
        
    # Below block of code will fetch the price
    temp_price1=soup.find_all('span',class_="a-price-whole")
    temp_price2=[]
    temp_price2.clear()
    for i in temp_price1:
        temp_price2.append(i.text)
    
    # Below block of code will fetch the img urls
    temp_pic1=soup.find_all('img' )
    temp_pic2=[]
    for i in temp_pic1:
        temp_pic2.append(i['src'])
    temp_pic3=[]
    temp_pic3.clear()
    temp_pic3=temp_pic2[3:]
    
    
    # Below block of code will fetch ratings
    temp_rating1=soup.find_all('span',class_="a-icon-alt")
    temp_rating2=[]
    for i in temp_rating1:
        temp_rating2.append(i.text.split(" ")[0])
    temp_rating3=[]
    temp_rating3.clear()
    temp_rating3=temp_rating2[0:-4]
    
    
    #Below block of code will create DataFrame
    mobile=pd.DataFrame({})
    mobile['Product Name']= temp_name2
    mobile['Price']= temp_price2
    mobile['Image']= temp_pic3
    mobile['Rating out of 5']= temp_rating3
    
    
    return mobile 
```

### Q.8 Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, SanFrancisco. You need to extract data about 7 day extended forecast display for the city.The data should include period, short description, temperature and description.

```python
def get_forecast(link):
    """this function will give weather forecast """
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    # Below Block will give period
    temp_period1=soup.find_all('div',class_="col-sm-2 forecast-label")
    period1=[]
    period1.clear()
    for i in temp_period1:
        period1.append(i.text)
    
    # Below block will give description
    temp_description1=soup.find_all('div',class_="col-sm-10 forecast-text")
    short_descr=[]
    short_descr.clear()
    for i in temp_description1:
        short_descr.append(i.text.split(',')[0])
        
    
    full_descr=[]
    full_descr.clear()
    for i in temp_description1:
        full_descr.append(i.text)
        
    # Below block will give temperatures
    temperature1=[]
    temperature1.clear()
    for i in temp_description1:
        temperature1.append(i.text.split(',')[1].split('.')[0])
    temperature1
    temperature2=[]
    temperature2.clear()
    for i in temperature1:
        temperature2.append(i[-2:])

    # Below block will create DataFrame
    SF_forecast=pd.DataFrame({})
    SF_forecast['Period']=period1
    SF_forecast['Weather']=short_descr
    SF_forecast['Temp in (degree F)']=temperature2
    SF_forecast['Description']=full_descr
    
    return SF_forecast 
```

```python
get_forecast('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996')
```
